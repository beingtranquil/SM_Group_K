```python
from scipy.stats import pearsonr
import numpy as np
import matplotlib.pyplot as plt

# class            wmc    SLOC
# ViewSalesPage     1       93
# Metricstics       1       31
# HandleData        15      36
# LoginPage         3       19


logical_sloc = [93, 31, 36, 19]
wmc = [1, 1, 15, 3]

# Calculate Pearson correlation coefficient
correlation_coefficient, _ = pearsonr(logical_sloc, wmc)
print(f'Correlation Coefficient: {correlation_coefficient:.2f}')


# Interpretation of the correlation coefficient
if correlation_coefficient > 0:
    print('Positive correlation')
elif correlation_coefficient < 0:
    print('Negative correlation')
else:
    print('No correlation')
# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(logical_sloc, wmc, color='blue')
plt.title('Scatter Plot of Logical SLOC vs WMC for METRICSTICS')
plt.xlabel('Logical SLOC')
plt.ylabel('WMC')
plt.grid(True)
plt.figtext(0.16, 0.70, f'Correlation Coefficient: {correlation_coefficient:.2f}', color='green', weight='bold')

plt.show()


```
